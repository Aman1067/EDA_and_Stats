```python
dg=df.copy()
```

```python
dg.head(10)
```

```python
dg=dg.drop(dg.index[10472])
```

```python
dg['Size']=dg['Size'].str.replace('M','000')
```

```python
dg['Size']=dg['Size'].str.replace('k','')
```

```python
dg.tail()
```

```python
dg['Size']=dg['Size'].str.replace('Varies with device','NaN')
```

```python
dg['Size']=dg['Size'].astype('float')
```

```python
dg['Size']=dg['Size'].str.replace(',','').str.replace('+','')
```

```python
dg['Size'].head(10)
```

```python
dg['Size'][2]*1000
```

```python
for i in dg['Size']:
    if i<10:
        dg['Size']=dg['Size'].replace(i,i*1000)
```

```python
dg['Size'].replace(8.7,8.7*1000)
```

```python
dg['Installs']=dg['Installs'].str.replace('+',"").str.replace(',','')
```

```python
dg['Reviews']=dg['Reviews'].astype('int')
```

```python
dg.info()
```

```python
pd.reset_option('display.max_rows')
```

```python
dg['Price']=dg['Price'].str.replace('$','')
```

```python
dg['Price']=dg['Price'].astype('float')
```

```python
dg['Installs']=dg['Installs'].astype('int')
```

```python
dg['Last Updated']=pd.to_datetime(dg['Last Updated'])
```

```python
dg.info()
```

```python
dg['Day']=dg['Last Updated'].dt.day
```

```python
dg['Month']=dg['Last Updated'].dt.month
```

```python
dg['Year']=dg['Last Updated'].dt.year
```

```python
dg.head()
```

```python
dg.to_csv('Playstore_cleaned_data.csv',index=False)
```
